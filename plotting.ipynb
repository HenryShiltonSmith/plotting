```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv = pd.read_csv('Employee_Int_Convert.csv')

# plt.title('Leave Or Not Histogram')
# plt.xlabel('LeaveOrNot (0 Leave)')
# plt.ylabel('Number of Records')
# csv['LeaveOrNot'].hist()

corr = csv.iloc[:,:-1].corr(method="pearson")
cmap = sns.diverging_palette(250,354,80,60,center='dark',as_cmap=True)
sns.heatmap(corr, vmax=1, vmin=-.5, cmap=cmap, square=True, linewidths=.2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.metrics import mean_squared_error, r2_score

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

csv = pd.read_csv('Employee_Int_Convert.csv')

X = csv[['Age']]
y = csv['LeaveOrNot']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
y_pred = lr.predict(X_test)

plt.scatter(X_test.iloc[:, 0], y_test, color='black')
# plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.title('Age vs LeaveOrNot')
plt.xlabel('Age')
plt.ylabel('LeaveOrNot')
plt.show()
```
